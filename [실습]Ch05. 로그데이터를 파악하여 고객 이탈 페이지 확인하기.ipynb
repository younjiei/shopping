#### 어떤 페이지에서 고객이 이탈을 할까?

#### 퍼널 스텝 dataframe 생성
 - 스텝 순서(ordering) 등을 명시하기 위해 사용

````python
funnel_dict={'/product_list':1,'/product_detail':2,'/cart':3,'/order_complete':4}
funnel_steps=pd.DataFrame.from_dict(funnel_dict,orient='index',columns=['step_no'])
funnel_steps
````

#### session, url 로 grouping
 - user_id가 아닌 session을 기준으로 삼는 이유는 **동일한 유저가 다른 세션으로 접속한 경우도 다른 경우로 간주**해야 하기 때문
 - session_id와 url로 그루핑하여 가장 시간대가 빠른 해당 이벤트에 대해 추출
 
 - ex)list로 갔다가 detail로 갔다 다시 list로 오는 상황을 생각하면 list log 중 가장 빠른 시간대

````python
grouped=logs.groupby(['session_id','url'])['datetime'].agg(np.min) #가장 빠른 이벤트로
grouped=pd.DataFrame(grouped).merge(funnel_steps,left_on='url',right_index=True)
grouped
````

#### 퍼널 테이블 생성
 - 각 퍼널의 스텝이 순서대로 columns으로 오도록 변경

````python
funnel=grouped.reset_index().pivot(index='session_id',columns='step_no',values='datetime')
funnel.columns=funnel_steps.index
funnel.head()
````

#### 퍼널 카운트 계산
 - 각 퍼널 스텝별 카운트 계산

````python
step_count=[funnel[i].notnull().sum() for i in funnel_steps.index]
step_count
````

````python
from plotly import graph_objects as go
def show_funnel(step_count,funnel_steps):

  fig=go.Figure(go.Funnel(
      x=step_count,
      y=funnel_steps
  ))
  fig.show()
````

````python
show_funnel(step_count,funnel_steps.index)
#결제에 문제가 있구나
````

#### 평균 시간 계산
 - 각 퍼널별 소요 시간 계산

````python
#/product_list	/product_detail
np.mean(funnel['/product_detail']-funnel['/product_list'])
#리스트를 보고 디테일에 진입하는데 평균 16분 소요
````

````python
#/product_detail	/cart
np.mean(funnel['/cart']-funnel['/product_detail'])
#디테일에서 카트 진입하는데 평균 약 18분 소요
````

````python
#/cart	/order_complete
np.mean(funnel['/order_complete']-funnel['/cart'])
#카트에서  구매완료까지 평균 약 33분 소요
#카트에서 결제까지 시간을 줄여야한다
````
